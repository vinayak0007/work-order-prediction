```python
def has_out_of_order(ts):
    timestamps = [x["timestamp"] for x in ts]
    return timestamps != sorted(timestamps)

df["out_of_order"] = df["timeseries_data"].apply(has_out_of_order)
df["out_of_order"].sum()
```

no out of order data

```python
def has_duplicates(ts):
    timestamps = [x["timestamp"] for x in ts]
    return len(timestamps) != len(set(timestamps))

df["has_duplicates"] = df["timeseries_data"].apply(has_duplicates)
df["has_duplicates"].sum()
```

no duplicates in timestamp

```python
def temp_bounds(ts):
    temps = [x["temperature"] for x in ts]
    return min(temps), max(temps)

bounds = df["timeseries_data"].apply(temp_bounds)

mins = [b[0] for b in bounds]
maxs = [b[1] for b in bounds]

print("Global min temp:", min(mins))
print("Global max temp:", max(maxs))
```

min temp & max temp are well within range

```python
#evaluating and plotting against temperature

import pandas as pd
import matplotlib.pyplot as plt

sample_wo = df[df["label"] == "WO"].iloc[0]["timeseries_data"]
sample_no = df[df["label"] == "NO_WORK_ORDER"].iloc[0]["timeseries_data"]

# Convert to DataFrame
wo_df = pd.DataFrame(sample_wo).sort_values("timestamp")
no_df = pd.DataFrame(sample_no).sort_values("timestamp")

# Convert to datetime
wo_df["timestamp"] = pd.to_datetime(wo_df["timestamp"], unit="s")
no_df["timestamp"] = pd.to_datetime(no_df["timestamp"], unit="s")

# Normalize time (relative comparison)
wo_df["time_since_start_min"] = (
    (wo_df["timestamp"] - wo_df["timestamp"].min())
    .dt.total_seconds() / 60
)

no_df["time_since_start_min"] = (
    (no_df["timestamp"] - no_df["timestamp"].min())
    .dt.total_seconds() / 60
)

# Plot
plt.figure(figsize=(10, 5))

plt.plot(
    wo_df["time_since_start_min"],
    wo_df["temperature"],
    label="WO"
)

plt.plot(
    no_df["time_since_start_min"],
    no_df["temperature"],
    label="NO_WORK_ORDER"
)

plt.xlabel("Time Since Start (minutes)")
plt.ylabel("Temperature")
plt.title("Temperature Pattern Comparison: WO vs NO_WORK_ORDER")
plt.legend()

plt.show()
```

major signal: no_work_order has stabler temperature variation whereas for wo, there is a lot of variation in temperature

```python
#extracting features out of temperature values in time series data

import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis
from sklearn.linear_model import LinearRegression


def compute_features(timeseries):
    ts_df = pd.DataFrame(timeseries)
    ts_df = ts_df.sort_values("timestamp")

    timestamps = ts_df["timestamp"].values
    temps = ts_df["temperature"].values

    features = {}

    features["mean_temp"] = np.mean(temps)
    features["std_temp"] = np.std(temps)
    features["min_temp"] = np.min(temps)
    features["max_temp"] = np.max(temps)
    features["range_temp"] = features["max_temp"] - features["min_temp"]
    features["skew_temp"] = skew(temps)
    features["kurtosis_temp"] = kurtosis(temps)

    features["duration"] = timestamps[-1] - timestamps[0]
    features["num_points"] = len(temps)

    if len(temps) > 1:
        X = timestamps.reshape(-1, 1)
        model = LinearRegression().fit(X, temps)
        features["trend_slope"] = model.coef_[0]
        features["trend_r2"] = model.score(X, temps)
    else:
        features["trend_slope"] = 0
        features["trend_r2"] = 0

    return features
```

```python
feature_df = df["timeseries_data"].apply(compute_features)
feature_df = pd.DataFrame(feature_df.tolist())
feature_df["label"] = df["label"].map({"WO":1, "NO_WORK_ORDER":0})
```

```python
feature_df.groupby("label")[["std_temp", "range_temp"]].describe()
```

## ML model building 

```python
feature_df = df["timeseries_data"].apply(compute_features)
feature_df = pd.DataFrame(feature_df.tolist())
feature_df["label"] = df["label"].map({"WO":1, "NO_WORK_ORDER":0})

X = feature_df.drop(columns=["label"])
y = feature_df["label"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

lr_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(class_weight="balanced"))
])

lr_pipeline.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    class_weight="balanced",
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

lr_scores = cross_val_score(lr_pipeline, X, y, cv=cv, scoring="f1")
rf_scores = cross_val_score(rf, X, y, cv=cv, scoring="f1")

print("LR F1:", lr_scores.mean())
print("RF F1:", rf_scores.mean())
```

too high F1 values, will need further cross checking

```python
from sklearn.metrics import classification_report, roc_auc_score

for name, model in {
    "Logistic Regression": lr_pipeline,
    "Random Forest": rf
}.items():
    
    preds = model.predict(X_test)
    probs = model.predict_proba(X_test)[:, 1]
    
    print(f"\n{name}")
    print(classification_report(y_test, preds))
    print("ROC-AUC:", roc_auc_score(y_test, probs))
```

very idealistic values for precision, recall and F1, will need cross checking

```python
import pandas as pd

importances = rf.feature_importances_

importance_df = pd.DataFrame({
    "feature": X.columns,
    "importance": importances
}).sort_values("importance", ascending=False)

importance_df
```

The feature importance distribution shows that classification is primarily driven by volatility-based features such as range and standard deviation. Temporal and trend-based features contribute negligibly, indicating the problem is fundamentally a spread-separation task rather than a sequential modeling problem.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

for col in X.columns:
    model = LogisticRegression()
    model.fit(X_train[[col]], y_train)
    preds = model.predict_proba(X_test[[col]])[:, 1]
    auc = roc_auc_score(y_test, preds)
    print(col, round(auc, 3))
```

evaluating the discriminative power of each feature individually by fitting a univariate logistic regression model and measuring ROC-AUC. This helps identify which features independently separate the classes.

std_temp = 1.0
range_temp = 1.0

shows that there is literally no overlap between classes in these dimensions.

```python
import numpy as np

y_shuffled = np.random.permutation(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_shuffled, test_size=0.2, stratify=y_shuffled, random_state=42
)

lr_pipeline.fit(X_train, y_train)
print(lr_pipeline.score(X_test, y_test))
```

Validating the pipeline using a label-shuffling sanity check. After randomizing the labels, model accuracy dropped to ~0.5, confirming there is no data leakage. The high original accuracy is due to genuine separability of volatility-based features.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.heatmap(X.corr(), annot=True, cmap="coolwarm")
plt.show()
```

The heatmap reveals very strong positive correlations among dispersion-based features such as std_temp, range_temp, max_temp, and min_temp. This indicates high multicollinearity, as these variables are capturing the same underlying volatility signal. Since range_temp is mathematically derived from max_temp and min_temp, and strongly associated with std_temp, these features are largely redundant. This confirms that the classification task is primarily driven by temperature variability rather than temporal trend or coverage characteristics.
